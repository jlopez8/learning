You are now going to try to find the best rate of savings to achieve a down payment on a 1M house in 36 months. Since hitting this exactly is a challenge, we simply want your savings to be within 100 of the required down payment.
In ps1c.py, write a program to calculate the best savings rate, as a function of your starting salary. You should use bisection search to help you do this efficiently. You should keep track of the number of steps it takes your bisections search to finish. You should be able to reuse some of the code you wrote for part B in this problem.
Because we are searching for a value that is in principle a float, we are going to limit ourselves to two decimals of accuracy (i.e., we may want to save at 7.04\%  or 0.0704 in decimal â€“ but we are not going to worry about the difference between 7.041\% and 7.039\%). This means we can search for an integer between 0 and 10000 (using integer division), and then convert it to a decimal percentage (using float division) to use when we are calculating the current_savings after 36 months. By using this range, there are only a finite number of numbers that we are searching over, as opposed to the infinite number of decimals between 0 and 1. This range will help prevent infinite loops. The reason we use 0 to 10000 is to account for two additional decimal places in the range 0\% to 100\%. Your code should print out a decimal (e.g. 0.0704 for 7.04\%).

```python
# begin code....

# pseudo-global variables
total_cost = 1000000 # dream house cost
portion_down_payment = 0.25 # portion of downpayment cost
r = 0.04 # annual rate of return
semi_annual_raise = 0.07 # active every 6 months

# months to save
months_to_save = 36 # vriable not used
```

```python
def annual_salary_mod(months_to_save, annual_salary):
    # ====================== YOUR CODE HERE ======================
    
    # increase the annual salary after every 6th month
    if( months_to_save%6 == 0 and months_to_save>0 ):
        annual_salary = annual_salary*(1 + semi_annual_raise)
    # =============================================================
    return annual_salary
```

```python
def monthly_return(current_savings, r):
    """
    Compute monthly return given current savings and annual rate.
    
    Parameters
    ----------
    current_savings : float
        The input current_savings as a decimal.
        In the future vectorize.
        
    r : float
        The input annual return rate as a decimal.
        In the future vectorize.
    
    Returns
    -------
    monthly_return : float
        The computed contribution of the annual return rate based on current savings.
        
    Instructions
    ------------

    """

    # ====================== YOUR CODE HERE ======================
    monthly_return = current_savings*(r/12.0);
    
    # =============================================================
    return monthly_return
```

```python
def saved_after_36mo(
    annual_salary_initial, semi_annual_raise, 
    portion_saved, total_cost, 
    portion_down_payment, r):
    """
    Compute monthly return given current savings and annual rate.
    
    Parameters
    ----------
    annual_salary : float
        The input annual_salary as a decimal.
        In the future vectorize.
        
    portion_saved : float
        The input portion to save per mo. of annual salary.
        In the future vectorize.
        
    total_cost : float
        The input total dream house cost.
        In the future vectorize
    
    portion_down_payment : float  
        The input portion of total cost representing downpayment.
        Is a decimal. 
        
    semi_annual_raise : float
        The input semi_annual_raise as a decimal.
        In the future vectorize. 
    
    r : float
        The input annual percent rate as a decimal
    
    
    Returns
    -------
    current_savings : float
        The computed amount saved after 36 months.
        
    Instructions
    ------------

    """

    # ====================== YOUR CODE HERE ======================
    # At the end of each month, your savings will be increased by the return on your investment,
    # plus a percentage of your monthly salary (annual salary / 12).
 
    current_savings = 0  
    
     #save for 36 months 
    for idx_mo in range(37):
        # increase the annual salary after every 6th month
        annual_salary = annual_salary_mod(months_to_save, annual_salary_initial)
        
        # ROI interest
        ROI = monthly_return(current_savings, r) 
        
        # put a pin in this b/c it's dependent on a float
        # we may be more interested in a variable 
        salary_contribution = portion_saved*(annual_salary/12)
        
        # update savings
        current_savings += ROI + salary_contribution
    # =============================================================
    
    return current_savings
```

```python
def bisection_search(search_list, 
                     portion_down_payment, 
                     annual_salary_initial, 
                     total_cost):
    
    """
    x : integer (depracated)
    The input value to find in the list. 
    We don't know what value to look for.
    
    search_list : array_like integers. 
    The input array ordered where to find x.
    
    note - this assumes that x is in the search_list
    
    """
    
    # ====================== YOUR CODE HERE ======================
    MAX_ITERATIONS = 100000
    iteration_counter = 1
    left = 0
    right = len(search_list) - 1
    mid = (left + right)//2 #floored division
    
    # perform serch 
    goal_pay = portion_down_payment*total_cost
    
    
    #update condition to reflect our situation

    #initialize initial portion with the mid value
    #portion_saved = mid/10000 
    portion_saved = 0
    saved_after_36mo_ = 0
    
    while saved_after_36mo_ < goal_pay - 100 or saved_after_36mo_ > goal_pay + 100:
        
        saved_after_36mo_ = saved_after_36mo(
            annual_salary_initial, semi_annual_raise, 
            portion_saved, total_cost, 
            portion_down_payment, r)
                     
        if saved_after_36mo_ < goal_pay:
            # update the left boundary to be one more than the mid
            left =  mid + 1
        else:
            # update the right boundary to be one less than the mid
            right = mid - 1
        
        # update the mid
        mid = (left + right)//2
        
        # update to the new portion saved
        portion_saved = mid/10000
        
        # increment the iterations
        iteration_counter += 1
        
        if( iteration_counter > 5002 ):
            break
    
    # print('bisection method iterations = ', iterations)
            
    # =============================================================
    return portion_saved, iteration_counter
```

```python
def optimal_savings_proportion(
    annual_salary_initial, 
    semi_annual_raise, 
    total_cost, 
    portion_down_payment, 
    r):
    
    numerator_search_range = range(10001)
    
    goal_savings = total_cost*portion_down_payment
    
    months_to_save = 36 # this is unused
    
    portion_saved, iteration_counter = bisection_search(
        numerator_search_range, portion_down_payment,  annual_salary_initial, total_cost)
        
    return portion_saved, iteration_counter
```
